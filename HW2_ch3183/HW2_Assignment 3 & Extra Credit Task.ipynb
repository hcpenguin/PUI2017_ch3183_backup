```python
#use pandas read data
data2=pd.read_csv(os.getenv("DFDATA")+'/dvm6-udsa/1473437627/dvm6-udsa.csv',nrows=500)
```

```python
data2.head()
```

```python
#remove NaN data and filter out the columns are needed
data2.dropna(axis=0,how="all")
reduced_data2 = data2[["Violation Date","Charge #1: Infraction Amount"]]
reduced_data2.head()
```

```python
#Change Columns Names
reduced_data2.columns=["Date","Charge"]
reduced_data2.head()
```

```python
#Strip out columns dollar sign and change data type
reduced_data2["Charge"] = reduced_data2["Charge"].replace('[$]','',regex =  True).astype(float)
reduced_data2.head()
```

```python
#test data type
reduced_data2.dtypes
```

```python
#Change timeline data type
reduced_data2["Date"] = pd.to_datetime(reduced_data2["Date"],format="%m/%d/%Y")
type(reduced_data2["Date"][0])
```

```python
#Plotting
plt.figure(figsize=(15,15))
plt.title("Violation Charge",fontsize=16)
plt.scatter(reduced_data2["Date"].tolist(), reduced_data2["Charge"])
plt.xlabel("Date")
plt.ylabel("Charge")
plt.show()
```
